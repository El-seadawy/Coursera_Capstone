```python
import lxml.html as lh
import numpy as np
import pandas as pd
import requests
```

```python
wiki_page_url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
```

```python
page = requests.get(wiki_page_url)
```

```python
doc = lh.fromstring(page.content)
elements = doc.xpath('//tr')
column_names = []
rows_data = []
for t in elements[0]:
    name=t.text_content()
    name = name.replace('\n','')
    column_names.append(name)
for i in range(1,len(elements)-4):
    values = []
    for t in elements[i]:
        values.append(t.text_content().replace('\n',''))
    rows_data.append([val for val in values])
```

```python
df = pd.DataFrame(columns=column_names,data=(i for i in rows_data))
df.head(15)
```

```python
long_lat_df = pd.read_csv('https://cocl.us/Geospatial_data')
```

```python
long_lat_df.head()
```

```python
df['Longitude'] = 0 #Deafult long: 0
df['Latitude'] = 0 #Default lat:0
for p,long,lat in zip(long_lat_df['Postal Code'],long_lat_df['Latitude'],long_lat_df['Longitude']):
    df.loc[df['Postal Code'] == p,'Longitude'] = long
    df.loc[df['Postal Code'] == p,'Latitude'] = lat
df.head()
```
