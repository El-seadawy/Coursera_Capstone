```python
import lxml.html as lh
import numpy as np
import pandas as pd
import requests
```

```python
wiki_page_url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
```

```python
page = requests.get(wiki_page_url)
```

```python
doc = lh.fromstring(page.content)
elements = doc.xpath('//tr')
column_names = []
rows_data = []
for t in elements[0]:
    name=t.text_content()
    name = name.replace('\n','')
    column_names.append(name)
for i in range(1,len(elements)-4):
    values = []
    for t in elements[i]:
        values.append(t.text_content().replace('\n',''))
    rows_data.append([val for val in values])
```

```python
df = pd.DataFrame(columns=column_names,data=(i for i in rows_data))
df.head(15)
```
